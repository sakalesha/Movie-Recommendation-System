```python
# =========================================================
# # Movie Recommendation System Using Supervised ML (Classification)
# =========================================================


# IMPORT LIBRARIES

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

from xgboost import XGBClassifier   # NEW MODEL


# LOAD DATASET

df = pd.read_csv('/kaggle/input/movierecommendationsystemdataset/MovieRecommendationSystem.csv')
print("Dataset Loaded Successfully!")


# HEAD + SHAPE

print(f"Dataset Shape: {df.shape}")
print("\nColumns:\n", df.columns)
display(df.head())


# MISSING VALUES CHECK

missing = df.isnull().sum()
print("Missing Values:\n", missing[missing > 0])


# FEATURE SELECTION (Embedded Tree-Based)

df['Target'] = (df['Movie_Vote'] >= 7).astype(int)

numeric_cols = ['Movie_Budget', 'Movie_Popularity', 'Movie_Runtime']
df['Movie_Genre'] = df['Movie_Genre'].fillna('Unknown')
genre_dummies = df['Movie_Genre'].str.get_dummies(sep=' ')

X_full = pd.concat([df[numeric_cols], genre_dummies], axis=1)
y = df['Target']

imputer = SimpleImputer(strategy='median')
X_full[numeric_cols] = imputer.fit_transform(X_full[numeric_cols])

selector_estimator = RandomForestClassifier(n_estimators=200, random_state=42)
selector_estimator.fit(X_full, y)

sfm = SelectFromModel(selector_estimator, threshold="median", prefit=True)
X_selected = pd.DataFrame(sfm.transform(X_full), columns=X_full.columns[sfm.get_support()])

print("\nSelected Features:", list(X_selected.columns))


# EDA

# (a) Top 10 Genres
all_genres = df['Movie_Genre'].str.split(' ', expand=True).stack()
top_genres = all_genres.value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_genres.index, y=top_genres.values)
plt.title('Top 10 Movie Genres')
plt.xticks(rotation=45)
plt.show()

# (b) Popularity vs Revenue
df['Movie_Popularity'] = df['Movie_Popularity'].fillna(0)
df['Movie_Revenue'] = df['Movie_Revenue'].fillna(0)

plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Movie_Popularity'], y=df['Movie_Revenue'], alpha=0.7)
plt.title("Popularity vs Revenue")
plt.show()

# (c) Heatmap
numeric_cols_all = ['Movie_Budget','Movie_Popularity','Movie_Revenue','Movie_Runtime','Movie_Vote']

plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols_all].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# (d) Vote Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Movie_Vote'].fillna(df['Movie_Vote'].mean()), bins=20, kde=True)
plt.title("Movie Vote Distribution")
plt.show()


# TRAIN-TEST SPLIT

X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y, test_size=0.2, random_state=42
)


# THREE CLASSIFICATION MODELS (RF, GB, XGB)


# Model 1: Random Forest
rf_model = RandomForestClassifier(n_estimators=200, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

# Model 2: Gradient Boosting
gb_model = GradientBoostingClassifier(n_estimators=200, learning_rate=0.1, random_state=42)
gb_model.fit(X_train, y_train)
y_pred_gb = gb_model.predict(X_test)

# Model 3: XGBoost Classifier
xgb_model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42,
    eval_metric='logloss'
)

xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)


# CLASSIFICATION REPORTS + CONFUSION MATRIX HEATMAPS


def show_model_results(model_name, y_true, y_pred, cmap):
    print(f"\n========== {model_name} ==========")
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    print("\nClassification Report:\n")
    print(classification_report(y_true, y_pred))

    # Heatmap Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap=cmap)
    plt.title(f"{model_name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()


# ------------------ Random Forest ------------------
show_model_results("Random Forest Classifier", y_test, y_pred_rf, cmap="Blues")

# ------------------ Gradient Boosting ------------------
show_model_results("Gradient Boosting Classifier", y_test, y_pred_gb, cmap="Greens")

# ------------------ XGBoost ------------------
show_model_results("XGBoost Classifier", y_test, y_pred_xgb, cmap="Oranges")



# ACCURACY COMPARISON GRAPH

models = ["Random Forest", "Gradient Boosting", "XGBoost"]
accuracy_scores = [
    accuracy_score(y_test, y_pred_rf),
    accuracy_score(y_test, y_pred_gb),
    accuracy_score(y_test, y_pred_xgb)
]

plt.figure(figsize=(7,5))
plt.bar(models, accuracy_scores)
plt.ylabel("Accuracy Score")
plt.title("Model Accuracy Comparison")
plt.ylim(0,1)
plt.show()


# BEST MODEL DESCRIPTION

accuracy_rf = accuracy_scores[0]
accuracy_gb = accuracy_scores[1]
accuracy_xgb = accuracy_scores[2]

best_accuracy = max(accuracy_rf, accuracy_gb, accuracy_xgb)

if best_accuracy == accuracy_rf:
    best_model_name = "Random Forest Classifier"
    best_model = rf_model
elif best_accuracy == accuracy_gb:
    best_model_name = "Gradient Boosting Classifier"
    best_model = gb_model
else:
    best_model_name = "XGBoost Classifier"
    best_model = xgb_model

print("\n================= BEST MODEL =================")
print(f"Best Model: {best_model_name}")
print(f"Accuracy: {best_accuracy:.4f}")

print("\nWhy this model is best?")
print("✔ Achieved the highest accuracy among all three models.")
print("✔ Handles selected features effectively.")
print("✔ Learns non-linear patterns extremely well.")
print("✔ Generalizes best on Hit vs Flop prediction.\n")


# 📌 ROC–AUC CURVES FOR ALL MODELS

from sklearn.metrics import roc_curve, auc

plt.figure(figsize=(8,6))

# Random Forest
y_prob_rf = rf_model.predict_proba(X_test)[:,1]
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
auc_rf = auc(fpr_rf, tpr_rf)
plt.plot(fpr_rf, tpr_rf, label=f"Random Forest (AUC = {auc_rf:.3f})")

# Gradient Boosting
y_prob_gb = gb_model.predict_proba(X_test)[:,1]
fpr_gb, tpr_gb, _ = roc_curve(y_test, y_prob_gb)
auc_gb = auc(fpr_gb, tpr_gb)
plt.plot(fpr_gb, tpr_gb, label=f"Gradient Boosting (AUC = {auc_gb:.3f})")

# XGBoost
y_prob_xgb = xgb_model.predict_proba(X_test)[:,1]
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_prob_xgb)
auc_xgb = auc(fpr_xgb, tpr_xgb)
plt.plot(fpr_xgb, tpr_xgb, label=f"XGBoost (AUC = {auc_xgb:.3f})")

# Plot settings
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC–AUC Curve Comparison")
plt.legend()
plt.show()



# PREDICT TOP 10 HIT MOVIES


df_features = df.copy()
df_features[numeric_cols] = imputer.transform(df_features[numeric_cols])

df_features_selected = pd.DataFrame(
    sfm.transform(df_features[numeric_cols].join(genre_dummies)),
    columns=X_selected.columns
)

df['Predicted_Hit'] = best_model.predict(df_features_selected)

top_movies = df[df['Predicted_Hit'] == 1].sort_values('Movie_Vote', ascending=False).head(10)

print("\nTop 10 Predicted Hit Movies:")
display(top_movies[['Movie_Title','Movie_Genre','Movie_Vote','Predicted_Hit']])
```
